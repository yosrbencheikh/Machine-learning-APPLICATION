```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data import + preprocess
data = pd.read_csv('/home/yosr/Dataset-youtube/train/train_topics.csv', 
error_bad_lines=False)

#replacing special characters (specific to this dataset) with spaces
spec_chars = ["►","■","●",
              "•"]
for char in spec_chars:
    data['title'] = data['title'].str.replace(char, ' ')
    data['description'] = data['description'].str.replace(char, ' ')

#removing double white spaces
data['title'] = data['title'].str.split().str.join(" ")
data['description'] = data['description'].str.split().str.join(" ")
```

```python
import re
#replaced all the empty cells/descriptions with 'none'
for i in data.columns:
	data[i][data[i].apply(lambda i: True if re.search('^\s*$', str(i)) else False)]=None

#replace none with NaN
data.fillna(value=pd.np.nan, inplace=True)

#put an insignificnt stopword
data = data.replace(np.nan, value='this', regex=True)
```

```python
#create a joined column
data['title_description'] = data['title'] + '' + data['description']
```
